Este projeto utilizará a ferramenta PyGaze para conduzir um estudo de cores com rastreamento ocular (eye-tracking). A ideia é observar como os usuários interagem visualmente com diferentes estímulos visuais (imagens, padrões, cores), registrando:
- quanto tempo cada imagem ou estímulo é olhado (duração de fixação)
- quantas vezes o olhar retorna para uma imagem específica
- qual ordem ou sequência de atenção entre vários estímulos

Com essas medições será possível identificar quais cores ou regiões visuais atraem mais atenção, quais causam desconforto visual ou são menos visualizadas.

Além disso, o projeto vai incluir técnicas de Inteligência Artificial para analisar os dados coletados. Os modelos de aprendizado de máquina poderão ser utilizados para:
- classificar cores e estímulos em categorias como “alta atenção” e “baixa atenção”
- agrupar perfis de usuários que apresentam padrões semelhantes de comportamento visual

Inicialmente, o projeto não partirá de uma base de dados pré-existente, mas sim de dados originais coletados nos experimentos. Futuramente, dependendo do feedback do professor, poderá ser incorporada uma base de dados externa para ampliar as análises e melhorar o desempenho dos modelos de IA.
